```python
import pandas as pd
from os import listdir
from os.path import isfile, join
```

```python
mypath = 'bd/'
files = [f for f in listdir(mypath) if isfile(join(mypath, f))]
files
```

```python
dfs = []
for file in files:
    dfs.append(pd.read_csv('bd/'+file, encoding = 'latin1', sep=';'))
    
```

```python
dfALL = pd.DataFrame()
for df in dfs:
    aux = df[(df['NM_PROJETO'].isna() == False) & (df['AN_BASE'] == 2018)]
    aux = aux[['ID_PROJETO', 'NM_PROJETO', 'NM_AREA_CONCENTRACAO', 'NM_LINHA_PESQUISA']]
    dfALL = dfALL.append(aux, ignore_index=True)

dfALL.to_csv('database.csv')
#dfs[0][(dfs[0]['NM_PROJETO'].isna() == False) & (dfs[0]['AN_BASE'] == 2018)]
```
